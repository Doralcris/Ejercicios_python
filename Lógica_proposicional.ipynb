```python
#definicion de la proposicion original
calor = True

#negacion de la proposicion original
negacion = not calor

print(f"proposicion original, hace calor {calor}")
print(f"negacion de la proposicion, hace calor {negacion}")
```

**Conjunción (∧) - "Y" →**  Ambas deben ser verdaderas para que toda la oración sea verdadera.

```python
#definicion de las proposiciones

llueve = True
hace_frio = False
frase =llueve and hace_frio
print(frase)
```

**Disyunción (∨) - "O" →** Basta que una sea verdadera para que todo sea verdadero.

```python
#definicion de dos preposiciones
voy_al_cine = True
voy_a_comer = False
frase =voy_al_cine or voy_a_comer
print(frase)
```

```python
hay_gasolina = True
bateria_cargada = False

#usamos el AND
puede_arrancar =hay_gasolina and bateria_cargada
print(puede_arrancar)

#usamos el OR
tiene_energia =hay_gasolina or bateria_cargada
print(tiene_energia)

#usamos el NOT
tanque_vacio =not hay_gasolina
print(f"El esta vacio tanque_vacio)
```

```python
sabe_phyton = True
sabe_mate = False
experiencia_anios = 3

#usamos el AND
entra_al_curso =sabe_phyton and sabe_mate
print(entra_al_curso)

#usamos OR
entra_al_curso = sabe_phyton and (sabe_mate or experiencia_anios >=5)
print(entra_al_curso)





```
